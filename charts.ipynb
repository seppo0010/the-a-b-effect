```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('MethodsMaterials&Data/abE1/abE1.csv')
df
```

```python
df['condLabel'].value_counts()
```

```python
from matplotlib import cm
df['dv<=2'] = df['dv'] <= 2
cmap = cm.get_cmap('Set1')
ax = pd.DataFrame(df.groupby('condLabel').mean()['dv<=2'][['badge', 'poster', 'xbpl']] * 100).plot.bar(rot=0, figsize=(4, 5))
plt.xlabel("")
ax.set_xticklabels(['ID', 'Póster', 'A/B entre ID y Póster'])
ax.get_legend().remove()
plt.title('% que evalúa al caso inapropiado')
;
```

```python
df = pd.read_csv('MethodsMaterials&Data/abE14/abE14.csv', encoding = "ISO-8859-1")
df
```

```python
pd.concat((df['AB1Condition'], df['AB2Condition'])).value_counts()
```

```python
from matplotlib import cm

df_stacked = pd.concat((df[['AB1Condition', 'AB1Rating']], df[['AB2Condition', 'AB2Rating']].rename(columns={'AB2Condition': 'AB1Condition', 'AB2Rating': 'AB1Rating'}))).dropna()
df_stacked['AB1Rating<=2'] = df_stacked['AB1Rating'] <= 2
cmap = cm.get_cmap('Set1')
ax = pd.DataFrame(df_stacked.groupby('AB1Condition').mean()['AB1Rating<=2'][['DrugA', 'DrugB', 'DrugAB']] * 100).plot.bar(rot=0, figsize=(4, 5))
plt.xlabel("")
ax.set_xticklabels(['Droga A', 'Droga B', 'A/B entre ambas'])
ax.get_legend().remove()
plt.title('% que evalúa al caso inapropiado')
;
```
